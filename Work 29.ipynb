```python
#Q1:   計算標準常態分配，小於1的機率有多大?
```

```python
# library
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import math
import statistics
import seaborn as sns
```

```python
# Python實現常態分配
#1.設定使用參數
mu=0
sigma=1
x=np.linspace(-4, 4, 500)
# 2.產生 pdf圖
y = stats.norm.pdf(x,mu, sigma)
plt.plot(x, y, "b-", linewidth=2)
#plt.grid(True)
plt.title('pdf of Norma(0,1)')
plt.show()
```

```python
# 3.計算常態分配的累積機率 (cumulative density function)，pdf 的積分
cumsum_probs =  stats.norm.cdf(x,mu, sigma)
plt.ylabel('P(X<=x)')
plt.xlabel('x')
plt.title(' cdf of Nomal(0,1)')
plt.plot(x, cumsum_probs,"b-", linewidth=2)
plt.show()
print(cumsum_probs[10])
```

```python
#小於1
stats.norm.cdf(1,mu, sigma)-stats.norm.pdf(1,mu, sigma)
```

```python
#Q2:  計算標準常態分配，大於1，小於 -1 的機率有多大?
```

```python
#大於1
1-stats.norm.cdf(1,mu, sigma)
```

```python
#小於 -1 
stats.norm.pdf(-1,mu, sigma)-stats.norm.cdf(-1,mu, sigma)
```
