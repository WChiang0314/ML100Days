```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.random.normal(2, 1, 75)
y = 2 + 1.5 * x + np.random.normal(0, 2, 75)
```

```python
sns.displot(x)
sns.displot(y)
```

```python
sns.residplot(x=x,y=y,lowess=True,color='g')
```

```python
#作業目標(2) 
#(1)使用 distplot()使用簡單的規則來正確猜測預設情況下正確的數位,但嘗試更多或更少的 bin 可能會顯示資料中的其他特徵: 
#(2)有無kde對圖形分布的影響
```

```python
# bin: 指的是特徵值, 
# kde: on/off
# sns.distplot();

sns.displot(x,bins=100,kde=True,rug=False)
```

```python

sns.displot(y,bins=100,kde=True,rug=True)
```
